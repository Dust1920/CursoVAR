```python
import numpy as np
import matplotlib.pyplot as plt


def demand():
    p = 0.12
    x = np.random.geometric(p)
    return x

def loss():
    y = np.random.uniform()
    return y

def new():
    u = np.random.poisson(5)
    return u

x_0 = 200
max_cap = 1100

history = [x_0]

steps = 100

for x in range(steps):
    xt = min(x_0 - demand() - np.floor(loss()) * x_0 + new(), max_cap)
    if xt<0:
        xt = 0
    history.append(xt)
    x_0 = xt


print(history)
plt.plot(history)
```
